## 1. Ë®ìÁ∑¥„Éá„Éº„Çø (train.csv)

```python
# Ë®ìÁ∑¥„Éá„Éº„Çø„ÅÆË™≠„ÅøËæº„Åø
train_df = pd.read_csv('Titanic - Machine Learning from Disaster/titanic/train.csv')

print(f"üìä „Éá„Éº„Çø„Çµ„Ç§„Ç∫: {train_df.shape[0]}Ë°å √ó {train_df.shape[1]}Âàó")
print(f"üìã ÂàóÂêç: {list(train_df.columns)}")
print("\n" + "="*80)

# ÊúÄÂàù„ÅÆ10Ë°å„ÇíË°®Á§∫
train_df.head(10)
```

```python
# Âü∫Êú¨Áµ±Ë®àÊÉÖÂ†±
print("üìà Âü∫Êú¨Áµ±Ë®àÊÉÖÂ†±")
print("="*30)
train_df.describe()
```

```python
# Ê¨†ÊêçÂÄ§„ÅÆÁ¢∫Ë™ç
missing = train_df.isnull().sum()
print("‚ö†Ô∏è Ê¨†ÊêçÂÄ§„ÅÆÁ¢∫Ë™ç")
print("="*30)
for col, count in missing.items():
    if count > 0:
        percentage = (count / len(train_df)) * 100
        print(f"{col}: {count}‰ª∂ ({percentage:.1f}%)")
    else:
        print(f"{col}: Ê¨†Êêç„Å™„Åó")
```

## 2. „ÉÜ„Çπ„Éà„Éá„Éº„Çø (test.csv)

```python
# „ÉÜ„Çπ„Éà„Éá„Éº„Çø„ÅÆË™≠„ÅøËæº„Åø
test_df = pd.read_csv('Titanic - Machine Learning from Disaster/titanic/test.csv')

print(f"üìä „Éá„Éº„Çø„Çµ„Ç§„Ç∫: {test_df.shape[0]}Ë°å √ó {test_df.shape[1]}Âàó")
print(f"üìã ÂàóÂêç: {list(test_df.columns)}")
print("\n" + "="*80)

# ÊúÄÂàù„ÅÆ10Ë°å„ÇíË°®Á§∫
test_df.head(10)
```

```python
# „ÉÜ„Çπ„Éà„Éá„Éº„Çø„ÅÆÊ¨†ÊêçÂÄ§
missing_test = test_df.isnull().sum()
print("‚ö†Ô∏è „ÉÜ„Çπ„Éà„Éá„Éº„Çø„ÅÆÊ¨†ÊêçÂÄ§")
print("="*30)
for col, count in missing_test.items():
    if count > 0:
        percentage = (count / len(test_df)) * 100
        print(f"{col}: {count}‰ª∂ ({percentage:.1f}%)")
    else:
        print(f"{col}: Ê¨†Êêç„Å™„Åó")
```
